We've defined the `histograms` function below, which takes a table with columns `Age` and `Salary` and draws a histogram for each one. It uses bin widths of 1 year for `Age` and $1,000,000 for `Salary`.

```python
def histograms(t):
    ages = t.column('Age')
    salaries = t.column('Salary')/1000000
    t1 = t.drop('Salary').with_column('Salary', salaries)
    age_bins = np.arange(min(ages), max(ages) + 2, 1) 
    salary_bins = np.arange(min(salaries), max(salaries) + 1, 1)
    t1.hist('Age', bins=age_bins, unit='year')
    plt.title('Age distribution')
    t1.hist('Salary', bins=salary_bins, unit='million dollars')
    plt.title('Salary distribution') 
    
histograms(full_data)
print('Two histograms should be displayed below')
```

# Demo: Probability 

```python
#creating an array that contains all the possible diceroll outcomes 
dice = make_array(1,2,3,4,5,6)
dice
```

```python
#before starting any for-loop, we are creating two variables 'success' and 'trial'
success = 0      #will count the number of 4-dice outcomes that contain 'at least one 6'
trial = 0        #will count the number of different 4-dice outcomes are possible

#these for-loops will combine the four different outcomes from each of the four dice. 
for first_roll in dice:
    for second_roll in dice:
        for third_roll in dice:
            for fourth_roll in dice:
                
                
                trial = trial + 1     #to count the trial
                #to combine each dice roll outcome into an array
                four_die_outcome = make_array(fourth_roll,third_roll,second_roll,first_roll) 
                
                print('simulation:',four_die_outcome ,"// testing 6's", four_die_outcome == 6)
                
                #check to see if the array contains at least one 6
                if np.count_nonzero(four_die_outcome == 6) >= 1:
                    success = success + 1 
```

```python
trial 
```

```python
trial == 6**4
```

```python
#calculating the probability by counting the number of successes over the number of possible trials
success / trial
```
